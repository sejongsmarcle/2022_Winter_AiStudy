```python
import numpy as np

x=[2, 4 , 6 , 8]
y=[81, 93, 91, 97]
```

```python
# x와 y의 평균값
mx = np.mean(x)
my = np.mean(y)
mx, my
```

```python
# 기울기 공식의 분모
divisor = sum([ (mx - i)**2 for i in x ])
divisor
```

```python
# 기울기 공식의 분자
def top(x, mx, y, my):
  d = 0
  for i in range(len(x)):
    d += ( x[i] - mx) * ( y[i] - my)
  return d

dividend = top(x, mx, y , my)
dividend
```

```python
# a: 기울기 b: y절편
a = dividend / divisor
b = my - (mx*a)
```

```python
print("기울기 a =", a)
print("y절편 b =", b)
```
